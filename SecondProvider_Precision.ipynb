```python
col_list=[]
true_pre=[]
false_pre=[]
for i in imp_name:
    if len(sp_1patient[sp_1patient[i]!='0.0'])!=0:
        if len(sp_1patient[sp_1patient[i]!='0.0'].is_SecondProvider2.value_counts().reindex(sp_1patient.is_SecondProvider2.unique(), fill_value=0))==2:
            col_list.append(i)
            select_true=sp_1patient[sp_1patient[i]!='0.0'].is_SecondProvider2.value_counts().sort_index().reindex(sp_1patient.is_SecondProvider2.unique(), fill_value=0)
            true_pre.append(select_true[1]/(select_true[0]+select_true[1]))
            select_false=sp_1patient[sp_1patient[i]=='0.0'].is_SecondProvider2.value_counts().sort_index().reindex(sp_1patient.is_SecondProvider2.unique(), fill_value=0)
            false_pre.append(select_false[0]/(select_false[0]+select_false[1]))
```

```python
sp_dict = {'Impresion_columns': col_list, 'True_Precision': true_pre, 'False_Precision': false_pre}
sp_precision_df = pd.DataFrame(sp_dict)
sp_precision_df.sort_values(by='True_Precision',ascending=False)#, ascending=True)
```
