# Second Provider

```python
col_list=[]
true_pre=[]
count_sp=[]
total_count=[]
for i in new_column:
    if len(sp1[sp1[i]!='0.0'])!=0:
        if len(sp1[sp1[i]!='0.0'].is_SecondProvider2.value_counts().reindex(sp1.is_SecondProvider2.unique(), fill_value=0))==2:
            col_list.append(i)
            select_true=sp1[sp1[i]!='0.0'].is_SecondProvider2.value_counts().sort_index().reindex(sp1.is_SecondProvider2.unique(), fill_value=0)
            true_pre.append(select_true[1]/(select_true[0]+select_true[1]))
            count_sp.append(select_true[1])
            total_count.append(select_true[0]+select_true[1])
```

```python
sp_dict = {'Col_list': col_list, 'True_Precision': true_pre,'Count_SecondProvider':count_sp,'Total_count':total_count}
sp_precision_df = pd.DataFrame(sp_dict)
sp_precision_df.sort_values(by='True_Precision',ascending=False)
```

```python
sp_precision_df.to_csv('secondprovider_unit&age.csv')
```
