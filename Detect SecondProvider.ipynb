```python
new_incident=incident[['PrimaryKey','CallConfirmedDT']]
new_incident
```

```python
res_unit=unit[['PrimaryKey','ResponseUnitID']]
res_unit
```

```python
test=res_unit.merge(new_incident,on='PrimaryKey')
test2=test.merge(new_patient,on='PrimaryKey')
test2
```

```python
notlevel1=test2.loc[(test2['TransportLOC']!='Level 1: 3 or More Providers (ALS Care)')&
                  (test2['TransportLOC']!='Level 4: 1 Provider (BLS Care)')]
```

```python
list4 = [x[0] for x in notlevel1['ResponseUnitID']]
notlevel1['NewUnit']=list4
notlevel12=notlevel1.loc[(notlevel1['NewUnit']=='M')|(notlevel1['NewUnit']=='T')|(notlevel1['NewUnit']=='E')]
notlevel12=notlevel12.dropna()
```

```python
#removediff
outservice['diff time']=outservice['OOSInServiceDT']-outservice['OOSOutOfServiceDT']
removediff=outservice[outservice['diff time']!='00:00:00']
list5 = [x[0] for x in removediff['UnitID']]
removediff['NewUnit']=list5
removediff.loc[(removediff['NewUnit']=='M')|(removediff['NewUnit']=='T')|(removediff['NewUnit']=='E')]
```

```python
def time_in_range(start, end, x):
    """Return true if x is in the range [start, end]"""
    if start <= end:
        return start <= x <= end
    else:
        return start <= x or x <= end
```

```python
list1=[]
for i,j in zip(removediff['UnitID'],removediff['OOSOutOfServiceDT']):
    for a,b,c,d in zip(notlevel12['ResponseUnitID'],notlevel12['CallConfirmedDT']
                       ,notlevel12['UnitTransportedDT'],notlevel12['PrimaryKey']):
        if i==a:
            time_in_range(b,c,j)
            if time_in_range(b,c,j)==True:
                list1.append(d)
```

```python
list2=[]
for i,j in zip(removediff['UnitID'],removediff['OOSInServiceDT']):
    for a,b,c,d in zip(notlevel12['ResponseUnitID'],notlevel12['CallConfirmedDT']
                       ,notlevel12['UnitTransportedDT'],notlevel12['PrimaryKey']):
        if i==a:
            time_in_range(b,c,j)
            if time_in_range(b,c,j)==True:
                list2.append(d)
```

```python
list1=list(set(list1))
list2=list(set(list2))
case3=[]
for i in list1:
    for j in list2:
        if i==j:
            case3.append(i)
len(case3)
```

```python
all_case=case1+case2+case3
all_case=list(set(all_case))
len(all_case)
```

```python
len(case1)+len(case2)
len(set(case1+case2))
```
